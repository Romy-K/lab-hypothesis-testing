```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
# First, create new dfs for Grass and Dragon 
grass_df = df[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Grass')]

# Dragon-type PokÃ©mon (either in Type 1 or Type 2)
dragon_df = df[(df['Type 1'] == 'Dragon') | (df['Type 2'] == 'Dragon')]
```

```python
grass_df.describe()
```

```python
dragon_df.describe()
```

```python
# Histogram for Grass HP 

sns.histplot(grass_df['HP'], kde=True, color="salmon")
plt.show()
```

```python
# Histogram for Dragon HP

sns.histplot(dragon_df['HP'], kde=True, color="blue")
plt.show()
```

```python
# Check distribution of both dfs, use Shapiro for smaller sample size

# H0 = Distribution is normal 
# H1 = Distribution is not normal

# Conducting the Shapiro-Wilk test for normality on Dragon HP
shapiro_stat_grass, shapiro_p_grass = stats.shapiro(grass_df['HP'])

shapiro_stat_grass, shapiro_p_grass 
# Distribution is close to normal, but not normal 
```

```python
# Conducting the Shapiro-Wilk test for normality on Dragon HP
shapiro_stat_dragon, shapiro_p_dragon = stats.shapiro(dragon_df['HP'])

shapiro_stat_dragon, shapiro_p_dragon
# Distribution is normal
```

```python
# Q-Q for extra visual aid - Grass

stats.probplot(grass_df['HP'], plot=plt)
plt.show()
```

```python
# Q-Q for extra visual aid - Dragon

stats.probplot(dragon_df['HP'], plot=plt)
plt.show()
```

```python
# I think they are normal enough to try t-testing. 
# First, check if variances are equal 

from scipy.stats import levene
stat, p = levene(dragon_df['HP'], grass_df['HP'])
print(f"Levene's test p-value: {p:.5f}")
```

```python
# Independent two-sample t-test with equal variance
t_stat, t_p = stats.ttest_ind(dragon_df['HP'], grass_df['HP'], equal_var=True)

print(f"T-test Statistic: {t_stat:.4f}")
print(f"P-value: {t_p:.5f}")

# H0 = There is no relation between dragon HP and grass HP 
# p = < 0,05, so we reject the hypothesis. There is a significant relationship between dragon HP and grass HP. 
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Hypothesis: there is a significant difference between legendary and non-legendary pokemon when it comes to their stats. 
```

```python
# Legendary column is a boolean.
print(df.Legendary.value_counts())
# Let's create separate groups first

legendary = df[df['Legendary'] == True]
non_legendary = df[df['Legendary'] == False]
```

```python
# Now, t-test all of the stat groups separately 

for stat in ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']:
    t_stat, p_val = ttest_ind(legendary[stat], non_legendary[stat], equal_var=False)
    print(f"{stat}: t={t_stat:.4f}, p={p_val:.5f}")

# We can reject the null hypothesis: there is a statistically significant different between the stats of legendary and non-legendary pokemon. 
```

```python
# I want to check them all in a bunch as well. We use MANOVA (and clean up the column names) 
from statsmodels.multivariate.manova import MANOVA

manova = MANOVA.from_formula('HP + Attack + Defense + Sp_Atk + Sp_Def + Speed ~ Legendary', data=df.rename(columns={'Sp. Atk': 'Sp_Atk', 'Sp. Def': 'Sp_Def'}))
print(manova.mv_test())

# Hypothesis confirmed: P values are 0, there is a statistically significant difference between Legendary and non-Legendary stats
```

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
school_coord = (-118, 34)
hospital_coord = (-122, 37)

# Function to calculate Euclidean distance
def euclidean_dist(lon1, lat1, lon2, lat2):
    return np.sqrt((lon2 - lon1)**2 + (lat2 - lat1)**2)

# Add new columns for distances
df['distance_to_school'] = df.apply(lambda row: euclidean_dist(row['longitude'], row['latitude'], *school_coord), axis=1)
df['distance_to_hospital'] = df.apply(lambda row: euclidean_dist(row['longitude'], row['latitude'], *hospital_coord), axis=1)

df
```

```python
df_close_houses = df[(df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)]
df_far_houses = df[(df['distance_to_school'] >= 0.50) & (df['distance_to_hospital'] >= 0.50)]
```

```python
df_close_houses 
```

```python
df_far_houses
```

```python
# Check if distribution is normal for both 

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.histplot(df_close_houses['median_house_value'], kde=True, color='blue', label='Close', stat='density')
sns.histplot(df_far_houses['median_house_value'], kde=True, color='red', label='Far', stat='density')

plt.title("Distribution of Median House Values")
plt.xlabel("Median House Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Distribution is not normal for both groups
# Use Mann-Whitney U test 
```

```python
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df_close_houses['median_house_value'], df_far_houses['median_house_value'], alternative='two-sided')
print(f"U-statistic = {stat:.2f}, p-value = {p:.5f}")
```

```python
# We can reject the null: there is a significant difference between the two groups. Our hypothesis is supported. 
```
